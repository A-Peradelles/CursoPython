### Classe Retangulo:

* Atributos: LadoA, LadoB
* Métodos: Mudar valor dos lados, Retornar valor dos lador, calcular Área e calcular Perímetro
* Crie um programa que utilize esta classe. Ele deve pedir os usuário que unforme as medidas de um local. Depois, deve criar um objeto com as medidar e calcular a quantidade de pisos e de rodapés necessárias para o local.

```python
class Retangulo:
  def __init__(self, LadoA, LadoB):
    self.LadoA=LadoA
    self.LadoB=LadoB
    
  def mudarLados(self,LadoA,LadoB):
    self.LadoA=int(LadoA)
    self.LadoB=LadoB
    
  def valorLados(self):
    print(self.LadoA, self.LadoB)
    
  def area(self):
    area=self.LadoA*self.LadoB
    print(area)
    
  def perimetro(self):
    perimetro=(2*self.LadoA)+(2*self.LadoB)
    print(perimetro)
    
#teste 
lado1=float(input("digite o valor de um dos lados:"))
lado2=float(input("digite o valor do outro lado:"))
ret1=Retangulo(lado1,lado2)
ret1.valorLados()
ret1.area()
ret1.perimetro()

print()
lado1=float(input("digite o valor de um dos lados:"))
lado2=float(input("digite o valor do outro lado:"))
ret1.mudarLados(lado1,lado2)
ret1.valorLados()
```

### Classe Pessoa:

* Atributos: nome, idade, peso e altura
* Métodos: Envelhecer, engordar, emagrecer, crescer. Obs: Por padrão, a cada ano que nossa pessoa envelhece, sendo a idade dela menor que 21 anos, ela deve crescer 0,5 cm

```python
class Pessoa:

  def __init__(self, nome, idade, peso, altura):
    self.nome=nome
    self.idade=idade
    self.peso=peso
    self.altura=altura
    self.cresceu=0
    
  def engordar(self,numero):
    self.peso=self.peso+numero
    print("novo peso é", self.peso)
    
  def emagrecer(self,numero):
    self.peso=self.peso-numero
    print("novo peso é", self.peso)
    
  def envelhecer(self,numero):
    for i in range (1,numero+1):
        if self.idade<21:
            self.altura+=0.5
            self.idade+=1
            self.cresceu+=0.5
        else:
            self.idade+=1
    print("nova idade é", self.idade)
    
  def crescer(self):
    print("Cresceu %f cm" %self.cresceu)
    
#teste 
pes1=Pessoa("André", 5, 60, 100)
print(pes1.nome, pes1.idade, pes1.peso, pes1.altura)
print()

pes1.engordar(0.5)
print()

pes1.emagrecer(0.7)
print()
pes1.crescer()
print()

pes1.envelhecer(2)
print("sua altura é",pes1.altura)
pes1.crescer()
```

### Classe Conta Corrente: Crie uma classe para implementar uma conta corrente. A classe deve possuir os seguintes atributos: número da conta, nome do corentista e saldo. Os métodos são os seguintes: alterarNome, depósito e saque; No construtor, saldo é opcional, com valor default zero e os demais atributos são obrigatórios.

```python
class Conta_Corrente:
  def __init__(self, conta, nome):
    self.Conta=conta
    self.Nome=nome
    self.Saldo=0
    
  def alterarNome(self,nome):
    self.Nome=nome
    print("O nome da conta foi alterado para", self.Nome, "\n")
    
  def deposito(self,valor):
    self.Saldo=self.Saldo+valor
    print("Seu saldo é", self.Saldo, "\n")
    
  def saque(self,valor):
    self.Saldo=self.Saldo-valor
    print("Seu saldo é", self.Saldo, "\n")
    
  def dados(self):
    print("Número da conta:", self.Conta)
    print("Nome:", self.Nome)
    print("Saldo:", self.Saldo, "\n")
    
#Teste01
conta1=Conta_Corrente(12345,"Andre")
conta1.dados()

#Teste02
conta1.alterarNome("Hugo")
conta1.deposito(1000)
conta1.saque(200)
conta1.dados()
```

### Classe Bomba de Combustível:

* Atributos: tipoCombustivel, valorLitro, quantidadeCombustivel
* Métodos: abastecerPorValor(), abastecerPorLitro(), alterarValor(), alterarCombustivel(), alterarQuantidadeCombustivel()

```python
class Bomba_Combustivel:
  def __init__(self, Combustivel, ValorLitro, Quantidade):
    self.Combustivel=Combustivel
    self.ValorLitro=ValorLitro
    self.Quantidade=Quantidade
    
  def abastecerPorValor(self,valor): #abastecer um veiculo, diminui combustível na bomba
    QuanEmLitros=valor/self.ValorLitro
    if QuanEmLitros<=self.Quantidade:
        self.Quantidade=self.Quantidade-QuanEmLitros
    else:                                         # se valor equivaler a mais que a bomba pode fornecer
        diferenca=QuanEmLitros-self.Quantidade
        valor=valor-(diferenca*self.ValorLitro)
        self.Quantidade=0
    print("Valor abastecido:", valor)
    print("Litros na Bomba:", self.Quantidade, "\n")
    
  def abastecerPorLitro(self,valor):
    if valor<=self.Quantidade:
        self.Quantidade=self.Quantidade-valor
    else:
        diferenca=valor-self.Quantidade
        self.Quantidade=0
        valor=valor-diferenca
    print("Litros abastecidos:", valor)
    print("Litros na Bomba:", self.Quantidade, "\n")
    
  def alterarValor(self,valor):
    self.ValorLitro=valor
    print("Valor do Litro de combustível foi alterado para:", self.ValorLitro, "\n")
    
  def alterarCombustivel(self, combustivel, valor): #cada combustível tem seu valor associado
    self.Combustivel=combustivel
    self.ValorLitro=valor
    print("O combustível na Bomba é", self.Combustivel)
    print("O valor do Litro é", self.ValorLitro, "\n")
    
  def alterarQuantidadeCombustivel(self,valor):
    self.Quantidade=valor
    print("A quantidade em Litros de combustível é", self.Quantidade)

#Teste01
bomba=Bomba_Combustivel("Gasolina", 7, 100)
bomba.abastecerPorValor(70)
bomba.abastecerPorValor(700)
print("."*70)

#Teste02
bomba=Bomba_Combustivel("Gasolina", 7, 100)
bomba.abastecerPorLitro(10)
bomba.abastecerPorLitro(100)
print("."*70)

#Teste03
bomba=Bomba_Combustivel("Gasolina", 7, 100)
bomba.alterarCombustivel("Diesel", 6)
bomba.alterarValor(5)
bomba.alterarQuantidadeCombustivel(200)
```
