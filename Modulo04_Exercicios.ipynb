### 1. Faça um programa que leia um arquivo texto contendo uma lista de endereços IP e gere um outro arquivo, contendo um relatório dos endereços IP válidos e inválidos.

```python
#Passo01 - Criar arquivo com Lista de IP
print("Criar arquivo com Lista de IP")
arquivo=open("Lista_IP.txt","w+")
arquivo.write("200.135.80.9\n")
arquivo.write("192.168.1.1\n")
arquivo.write("8.35.67.74\n")
arquivo.write("257.32.4.5\n")
arquivo.write("85.345.1.2\n")
arquivo.write("1.2.3.4\n")
arquivo.write("9.8.234.5\n")
arquivo.write("192.168.0.256\n")
arquivo.close()

arquivo=open("Lista_IP.txt","r")
print(arquivo.read(),"\n")
arquivo.close()
print("."*70)
```

```python

#Passo02 - transformar em lista
print("transformar em lista")
arquivo=open("Lista_IP.txt","r")
lista=arquivo.readlines()
print(lista,"\n")
arquivo.close()
print("."*70)

#Passo03 - retirar "\n"
print("retirar pular linha")
contador=0
for i in lista:
    if "\n" in i:
        i=i.replace("\n","")
        lista[contador]=i
    contador+=1
print(lista,"\n")
print("."*70)

#Passo04 - separar IP em 4 e testar para valores entre 0 e 255
print("separar IP em 4 e testar para valores entre 0 e 255")
lista_valida=[]
lista_invalido=[]
for j in lista:
    separado=j.split(".")
    print(separado)
    for k in separado:
        k=int(k)
        if k>0 and k<255:
            valido=True
        else:
            valido=False
            break
    if valido==True:
        lista_valida.append(j)
    else:
        lista_invalido.append(j)
print("\n", lista_valida,"\n")
print(lista_invalido,"\n")
print("."*70)

#Passo05 - devolver "\n"
print("devolver pular linha")
contador=0
for i in lista_valida:
    lista_valida[contador]=lista_valida[contador]+"\n"
    contador+=1
print(lista_valida,"\n")

contador=0
for i in lista_invalido:
    lista_invalido[contador]=lista_invalido[contador]+"\n"
    contador+=1
print(lista_invalido,"\n")
print("."*70)

#Passo06 - Adicionar título
print("Adicionar título")
lista_valida.insert(0,"Enderecos validos:\n")
lista_invalido.insert(0,"Enderecos invalidos:\n")
print(lista_valida,"\n")
print(lista_invalido,"\n")
print("."*70)

#Passo07 - transformar lista em String
print("transformar lista em String")
lista_v=str()
for i in lista_valida:
    lista_v+=str(i)
print(lista_v)
print(type(lista_v))

lista_inv=str()
for i in lista_invalido:
    lista_inv+=str(i)
print(lista_inv)
print(type(lista_inv))
print("."*70)

#Passo08 - Criar arquivo de final
print("Criar arquivo de final")
arquivo=open("Lista_IP2.txt","w+")
arquivo.write(lista_v)
arquivo.write("\n")
arquivo.write(lista_inv)
arquivo.close()
```

### 2. Quando falamos sobre erros e exceçoes, algumas fontes de informação consideram que no processo de depuração, podemos encontrar três tipos de erro: erro de sintaxe, erro de tempo de excução e erro semântico. defina o conceito de cada um deles

* Erro de sintaxe: 
    SintaxErro é um erro que ocorre quando se tenta chamar uma função mas tem erro na digitação, como falta de ":" ou (). (Para o programa)

* Erro de tempo de execução: 
    RuntimeError é quando ocorre um erro, mas esse erro não entra em uma categoria. Também conhecidos como "bugs". (Para o programa)

* Erro semântico: 
    SemanticError ou LogicError é um erro lógico, Quando se deseja fazer uma coisa mais por erro na escrita outra é realizada (sem aparentar ter avido erro, só conferindo o resultado). (Não para o programa)

```python
# 1.7 Escreva um programa para resolver equações do segundo grau representadas por ax2+bx+c usando a Fórmula de Bhaskara

#Passo 01 - Tratamento da entrada de dados
while True:
    try:
        a=float(input("Digite o valor de a: "))
        if a!=0:
            break
        else:
            print("a = 0, não é uma equação do segundo grau\n")
    except ValueError:
        print("Você não digitou um valor válido\n")

while True:
    try:
        b=float(input("Digite o valor de b: "))
        break
    except ValueError:
        print("Você não digitou um valor válido\n")
        
while True:
    try:
        c=float(input("Digite o valor de c: "))
        break
    except ValueError:
        print("Você não digitou um valor válido\n")
        

#Cálculo
delta=((b**2)-(4*a*c))
x1=((-b)+(delta**0.5))/(2*a)
x2=((-b)-(delta**0.5))/(2*a)

#Passo 02 - Tratamento de número complexo
try:
    print("As raizes são %f" %x1, "e %f" %x2)
except TypeError:
    print("A equação possui raiz complexa\n")
else:
    print("Programa conseguiu calcular\n")
finally:
    print("Programa finalizado\n")
```
